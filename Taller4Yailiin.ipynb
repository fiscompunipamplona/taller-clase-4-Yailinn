```python
from pylab import plot, show

y = [1., 2.4, 1.7, 0.3, 0.6, 1.8]

plot(y)
show()
```

```python
from numpy import linspace, sin, cos

x = linspace(0, 10, 1000)
y = sin(x)

plot(x,y)
show()
```

```python
from pylab import ylim, xlim, grid
x = linspace(0, 10, 1000)
y = sin(x)

plot(x,y)
ylim( -1.1, 1.1)
xlim(0., 10)
grid("on")
show()
```

```python
from pylab import xlabel, ylabel
x = linspace(0, 10, 1000)
y = sin(x)
plot(x,y)
ylim( -1.1, 1.1)
xlim(0., 10)
grid("on")
xlabel("Eje x")
ylabel("Eje y")
show()
```

```python
x = linspace(0, 10, 100)
y = sin(x)

plot(x,y,'ko')
show()
```

```python
y = sin(x)
plot(x,y,'k-')
y = cos(x)
plot(x,y,'k--')
grid("on")
show()
```

```python
from pylab import plot, show, xlabel, grid, ylabel
from numpy import loadtxt
plot(dataCO2[:,2], dataCO2[:,3],'g')
dataCO2 = loadtxt("co2_mm_gl.txt", float)
plot(dataCO2[:,2], dataCO2[:,4])
grid('on')
xlabel("Aos (tiempo)")
ylabel("Media Mensual-Tendencia (CO2)")
show()
```

A partir de la grfica, se puede decir que el CO2 en la atmosfera ha ido aumentado a medida que los aos van pasando. Se nota cloaramente que existe una relacin directa entre el tiempo y la cantidad de CO2, logrando que asimismo la cantidad de oxigeno en el ambiente sea menor.
